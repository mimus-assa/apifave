```python
import tensorflow as tf
import tensorflow.keras as keras
from tensorflow.keras import models
from tensorflow.keras import backend as k
from tensorflow.compat.v1 import ConfigProto
from tensorflow.compat.v1 import InteractiveSession
config = ConfigProto()
config.gpu_options.allow_growth = True
session = InteractiveSession(config=config)
import json
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing.image import load_img
import numpy as np
from os import listdir
from keras.preprocessing.image import img_to_array
from shutil import copyfile
```

```python
model_dir = 'apifave/models/dl/ocr_34ch_gray_t9708_20x40.h5'
model = models.load_model(model_dir)
with open('apifave/dict_ocr_33.json', 'r') as f:
    labels = json.load(f)
letters = list(labels)
characters = [l for l in listdir('apifave/images/letters_from_videos/a/')]
predos = []
for each_character in characters:
    img = load_img('apifave/images/letters_from_videos/a/'+each_character, color_mode = "grayscale", target_size = (20,40))
    img_array = img_to_array(img)
    img_array = img_array.reshape(1, 20, 40, 1).astype('float')
    results = model.predict_classes(img_array, batch_size=1)
    dst = 'apifave/images/letters_separated/'+letters[results[0]]+'/'+each_character
    copyfile('apifave/images/letters_from_videos/a/'+each_character, dst)
    #print(dst)
    print(each_character, str(letters[results[0]]))
    
#file_name =  (generator.filenames)
#results = model.predict_classes(generator, batch_size=1)

#result2 = [str(letters[np.where(result == np.amax(result))[1][0]])]
#classification_result.append(result2)


```
